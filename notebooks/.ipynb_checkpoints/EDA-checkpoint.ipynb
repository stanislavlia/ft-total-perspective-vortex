```python
#Run Type to Task Map
run_type_to_task = {
    "R01": {
        "name": "Baseline - Eyes Open",
        "task_type": "baseline",
        "labels": None
    },
    "R02": {
        "name": "Baseline - Eyes Closed",
        "task_type": "baseline",
        "labels": None
    },
    "R03": {
        "name": "Task 1 - Real Left/Right Fist",
        "task_type": "motor_execution",
        "paradigm": "left_right_hand",
        "labels": {
            "T1": "left_fist",
            "T2": "right_fist"
        }
    },
    "R04": {
        "name": "Task 2 - Imagine Left/Right Fist",
        "task_type": "motor_imagery",
        "paradigm": "left_right_hand",
        "labels": {
            "T1": "left_fist",
            "T2": "right_fist"
        }
    },
    "R05": {
        "name": "Task 3 - Real Fists/Feet",
        "task_type": "motor_execution",
        "paradigm": "hands_feet",
        "labels": {
            "T1": "both_fists",
            "T2": "both_feet"
        }
    },
    "R06": {
        "name": "Task 4 - Imagine Fists/Feet",
        "task_type": "motor_imagery",
        "paradigm": "hands_feet",
        "labels": {
            "T1": "both_fists",
            "T2": "both_feet"
        }
    },
    "R07": {
        "name": "Task 1 - Real Left/Right Fist",
        "task_type": "motor_execution",
        "paradigm": "left_right_hand",
        "labels": {
            "T1": "left_fist",
            "T2": "right_fist"
        }
    },
    "R08": {
        "name": "Task 2 - Imagine Left/Right Fist",
        "task_type": "motor_imagery",
        "paradigm": "left_right_hand",
        "labels": {
            "T1": "left_fist",
            "T2": "right_fist"
        }
    },
    "R09": {
        "name": "Task 3 - Real Fists/Feet",
        "task_type": "motor_execution",
        "paradigm": "hands_feet",
        "labels": {
            "T1": "both_fists",
            "T2": "both_feet"
        }
    },
    "R10": {
        "name": "Task 4 - Imagine Fists/Feet",
        "task_type": "motor_imagery",
        "paradigm": "hands_feet",
        "labels": {
            "T1": "both_fists",
            "T2": "both_feet"
        }
    },
    "R11": {
        "name": "Task 1 - Real Left/Right Fist",
        "task_type": "motor_execution",
        "paradigm": "left_right_hand",
        "labels": {
            "T1": "left_fist",
            "T2": "right_fist"
        }
    },
    "R12": {
        "name": "Task 2 - Imagine Left/Right Fist",
        "task_type": "motor_imagery",
        "paradigm": "left_right_hand",
        "labels": {
            "T1": "left_fist",
            "T2": "right_fist"
        }
    },
    "R13": {
        "name": "Task 3 - Real Fists/Feet",
        "task_type": "motor_execution",
        "paradigm": "hands_feet",
        "labels": {
            "T1": "both_fists",
            "T2": "both_feet"
        }
    },
    "R14": {
        "name": "Task 4 - Imagine Fists/Feet",
        "task_type": "motor_imagery",
        "paradigm": "hands_feet",
        "labels": {
            "T1": "both_fists",
            "T2": "both_feet"
        }
    }
}

MOTOR_CHANNELS = [
    'C3..',   # Left motor cortex (primary)
    'Cz..',   # Central motor area (feet)
    'C4..',   # Right motor cortex (primary)
    'Fc3.',   # Left frontal-central (premotor)
    'Fc4.',   # Right frontal-central (premotor)
    'Cp3.',   # Left central-parietal (sensorimotor)
    'Cp4.',   # Right central-parietal (sensorimotor)
    'C5..',   # Left lateral motor
    'C1..',   # Left medial motor
    'C2..',   # Right medial motor
    'C6..',   # Right lateral motor
    'Fc1.',   # Left medial frontal-central
    'Fc2.',   # Right medial frontal-central
    'Fc5.',   # Left lateral frontal-central
    'Fc6.',   # Right lateral frontal-central
    'Cp1.',   # Left medial central-parietal
    'Cp2.',   # Right medial central-parietal
    'Cp5.',   # Left lateral central-parietal
    'Cp6.'    # Right lateral central-parietal
]

def rename_annotations(raw, run_type):
    """
    Rename MNE annotations to readable task labels
    """
    task_info = run_type_to_task[run_type]
    
    if task_info['labels'] is not None:
        # Map T0, T1, T2 to readable names
        annotation_mapping = {
            'T0': 'rest',
            'T1': task_info['labels']['T1'],
            'T2': task_info['labels']['T2']
        }
        
        raw.annotations.rename(annotation_mapping)
    
    # Extract events with new names
    events, event_dict = mne.events_from_annotations(raw)
    
    return event_dict

def select_eeg_files_for_subject_by_paradigm(subject_id, paradigm):
    """
    Select motor imagery EEG files for a subject by paradigm
    """
    matching_runs = [
        run_id for run_id, info in run_type_to_task.items()
        if info.get('paradigm') == paradigm and info['task_type'] == 'motor_imagery'
    ]
    
    return [
        os.path.join(RAW_DATA_DIR, subject_id, f"{subject_id}{run_id}.edf")
        for run_id in sorted(matching_runs)
    ]


def load_and_concatenate_to_epochs(files, tmin=-0.5, tmax=3.5, picks=MOTOR_CHANNELS):
    """
    Load multiple EDF files and concatenate into single Epochs object
    """
    epochs_list = []
    
    for filepath in files:
        # Extract run_id
        run_id = extract_task_id(filepath)
        
        # Load raw
        raw = mne.io.read_raw_edf(filepath, preload=True, verbose=False)
        
        # Rename annotations
        event_dict = rename_annotations(raw, run_id)
        
        # Get events (exclude rest)
        events, _ = mne.events_from_annotations(raw)
        event_id = {k: v for k, v in event_dict.items() if k != 'rest'}
        
        # Create epochs
        epochs = mne.Epochs(
            raw, events, event_id=event_id,
            tmin=tmin, tmax=tmax,
            baseline=None,
            picks=picks,
            preload=True,
            verbose=False
        )
        
        epochs_list.append(epochs)
    
    # Concatenate
    return mne.concatenate_epochs(epochs_list)

```

### Epoching Time-Series
In EEG analysis,
epoching is the process of extracting specific time windows (segments) from a continuous EEG recording. These segments, called epochs, are typically used to isolate neural responses to specific stimuli or to break long resting-state recordings into manageable chunks.

### Computing Global PSD

Use records from all subjects to compare Power Spectrum Density for Motor Imagery Task (Left/Right Hand).

**This plot will show us how different PSD for task1 and task2 (left/right hand)**


```python
def compute_global_psd(paradigm='left_right_hand', n_subjects=109):
    """
    Compute average PSD across all subjects for a paradigm
    
    Returns:
    --------
    freqs : array
    psd_event1_global : array (n_channels, n_freqs)
    psd_event2_global : array (n_channels, n_freqs)
    event_names : tuple
    """
    
    all_psd_event1 = []
    all_psd_event2 = []
    
    for subject_num in tqdm(range(1, n_subjects + 1), desc="Processing subjects"):
        subject_id = f"S{subject_num:03d}"
        
        try:
            # Get files for this subject
            files = select_eeg_files_for_subject_by_paradigm(subject_id, paradigm)
            
            # Load epochs
            epochs = load_and_concatenate_to_epochs(
                files=files,
                tmin=TMIN,
                tmax=TMAX,
                picks=MOTOR_CHANNELS
            )
            
            # Apply bandpass filter
            epochs.filter(l_freq=8, h_freq=30, verbose=False)
            
            # Get event names
            event_names = list(epochs.event_id.keys())
            
            # Get data for each event
            data_event1 = epochs[event_names[0]].get_data()  # (n_trials, n_channels, n_times)
            data_event2 = epochs[event_names[1]].get_data()
            
            # Compute PSD for each trial and channel
            fs = epochs.info['sfreq']
            
            for trial in data_event1:
                trial_psd = []
                for ch_idx in range(trial.shape[0]):
                    freqs, psd = welch(trial[ch_idx], fs=fs, nperseg=256)
                    trial_psd.append(psd)
                all_psd_event1.append(trial_psd)
            
            for trial in data_event2:
                trial_psd = []
                for ch_idx in range(trial.shape[0]):
                    freqs, psd = welch(trial[ch_idx], fs=fs, nperseg=256)
                    trial_psd.append(psd)
                all_psd_event2.append(trial_psd)
                
        except Exception as e:
            print(f" Skipping {subject_id}: {e}")
            continue
    
    # Average across all trials and subjects
    psd_event1_global = np.mean(all_psd_event1, axis=0)  # (n_channels, n_freqs)
    psd_event2_global = np.mean(all_psd_event2, axis=0)
    
    return freqs, psd_event1_global, psd_event2_global, event_names
```

```python
freqs, psd_event1, psd_event2, event_names = compute_global_psd(TASK_PARAGIDM)
```

### Visualize PSD by task for each channel

```python
# Plot
channel_names = MOTOR_CHANNELS
n_channels = len(channel_names)

fig, axes = plt.subplots(n_channels, 1, figsize=(12, 3*n_channels))

for i, ch_name in enumerate(channel_names):
    ax = axes[i] if n_channels > 1 else axes
    
    ax.plot(freqs, psd_event1[i], label=event_names[0], linewidth=2)
    ax.plot(freqs, psd_event2[i], label=event_names[1], linewidth=2)
    
    ax.axvspan(8, 13, alpha=0.15, color='green', label='Mu')
    ax.axvspan(13, 30, alpha=0.15, color='orange', label='Beta')
    
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('PSD (V/Hz)')
    ax.set_title(f'Global PSD - {ch_name} (All Subjects)')
    ax.set_xlim(5, 30)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Averaged across all subjects: {event_names[0]} vs {event_names[1]}")
```
