
# EEG (Electroencephalography) - Simple Description


### What It Is
**EEG** = Recording electrical activity from the brain using sensors (electrodes) placed on the scalp.

### How It Works
- Neurons in the brain communicate using electrical signals
- When many neurons fire together, they create electrical fields
- These fields reach the scalp surface (weakened by skull/skin)
- Electrodes measure tiny voltage changes (microvolts, V)
- Signals are amplified and recorded over time

### What It Measures
**Voltage fluctuations** caused by synchronized brain activity, typically showing rhythmic patterns (brain waves) at different frequencies:
- **Alpha (8-12 Hz):** Relaxed, eyes closed
- **Beta (13-30 Hz):** Active thinking, alert
- **Mu (8-12 Hz):** Motor cortex at rest; suppressed during movement/motor imagery

### Why It's Useful
- **Non-invasive** (just stick electrodes on head, no surgery)
- **Real-time** (millisecond precision)
- **Safe and cheap** (compared to fMRI, invasive implants)
- Used for: BCIs, sleep studies, epilepsy diagnosis, cognitive research

### For this Project
EEG captures brain signals when subjects imagine hand/feet movements. Different body parts activate different brain regions, creating distinct voltage patterns we can classify with machine learning.

## Data Structure

### Data Structure Overview

```
data/raw/
 S001/                           Subject 001
    S001R01.edf                   14 runs (R01-R14)
    S001R01.edf.event             Each run has .edf + .edf.event
    S001R02.edf
    ...
 S002/                           Subject 002
 S003/                           Subject 003
 ...
 S109/                           Subject 109 (total 109 subjects)
 64_channel_sharbrough.pdf       Electrode placement diagram
 RECORDS                         Dataset metadata
 SHA256SUMS.txt                  File checksums
```

### Summary
- **109 subjects** (S001-S109)
- **14 runs per subject** (R01-R14)
- **2 files per run:** `.edf` (EEG data) + `.edf.event` (annotations)
- **Total:** 109  14  2 = **3,052 files**

### Key Files
- **EDF files:** 64-channel EEG recorded at 160 Hz
- **Event files:** T0 (rest), T1 (motion A), T2 (motion B) markers
- **Electrode map:** Shows sensor positions on scalp (10-10 system)

### Run Types Reference
| Run # | Task | Description | Classification Task |
|-------|------|-------------|---------------------|
| R01 | Baseline | Eyes open | - |
| R02 | Baseline | Eyes closed | - |
| **R03, R07, R11** | Task 1 | Real left/right fist | **T1=Left, T2=Right** |
| **R04, R08, R12** | Task 2 | Imagine left/right fist | **T1=Left, T2=Right** |
| **R05, R09, R13** | Task 3 | Real fists/feet | **T1=Fists, T2=Feet** |
| **R06, R10, R14** | Task 4 | Imagine fists/feet | **T1=Fists, T2=Feet** |


## EEG Frequency Bands Reference


### **Delta (0.5-4 Hz)**
- Deep sleep
- **Remove** (noise)

### **Theta (4-8 Hz)**
- Drowsiness, light sleep
- **Remove** (irrelevant)

### **Alpha (8-12 Hz)**
- Relaxed, eyes closed
- Occipital cortex (back of head)

### **Mu (8-12 Hz)** 
- Motor cortex at rest
- **Decreases during motor imagery (ERD)**
- **Primary BCI signal**

### **Beta (13-30 Hz)** 
- Active thinking, motor control
- **Decreases during motor imagery (ERD)**
- **Secondary BCI signal**

### **Gamma (30-100+ Hz)**
- High-level cognition
- **Not used** (too weak, artifacts)

## Raw data overview

````python
import numpy as np
import mne
import os
import matplotlib.pyplot as plt
import seaborn as sns
import glob
# Switch to matplotlib backend
mne.viz.set_browser_backend('matplotlib')

#=============SETTINGS============
RAW_DATA_DIR = "../data/raw"
SAMPLE_FILE=os.path.join(RAW_DATA_DIR, "S001/S001R02.edf")
````

````python
raw_eeg_data_sample = mne.io.read_raw_edf(SAMPLE_FILE)
events = mne.events_from_annotations(raw_eeg_data_sample)

raw_eeg_data_sample
````

## Visualize EEG Data Sample (EEG Viewer)

**Y-axis:** Channel names (64 electrodes on scalp)  
**X-axis:** Time (seconds)  
**Traces:** Each channel's voltage over time (black wavy lines)  
**Width of trace:** Voltage amplitude (wider = higher voltage)

**What it shows:** Raw, unfiltered EEG brain activity from all 64 sensors simultaneously. Currently very noisy - needs filtering (8-30 Hz bandpass) to reveal actual brain rhythms and motor imagery patterns.

**Key channels for motor imagery:** C3 (left motor cortex), C4 (right motor cortex), Cz (central)


````python
eeg_viewer_plot = raw_eeg_data_sample.plot(title="")
````

````python
s1_baseline_data = mne.io.read_raw_edf(os.path.join(RAW_DATA_DIR, "S001/S001R01.edf"))
s1_real_left_right_fist_data = mne.io.read_raw_edf(os.path.join(RAW_DATA_DIR, "S001/S001R03.edf"))
s1_img_left_right_fist_data = mne.io.read_raw_edf(os.path.join(RAW_DATA_DIR, "S001/S001R04.edf"))
````

````python


s1_baseline_data.plot(
    title='S001R01 - Baseline (Eyes Open)',
    duration=40,
    n_channels=10,
    scalings=dict(eeg=50e-6)
)

s1_real_left_right_fist_data.plot(
    title='S001R03 - Real Left/Right Fist',
    duration=40,
    n_channels=30,
    scalings=dict(eeg=50e-6)
)

s1_img_left_right_fist_data.plot(
    title='S001R04 - Imagine Left/Right Fist',
    duration=40,
    n_channels=30,
    scalings=dict(eeg=50e-6)
)
print()
````
