```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Create time-domain signal

```python
fs = 120 #sampling frequency
T = 1 / fs #sampling period
N = 10 #length of signal
t = np.arange(0, 100) #time vector

#construct signal
signal = 6 * np.sin(2*np.pi*0.4 * t) - 10 * np.cos(2*np.pi*0.1 * t) + np.random.randn(len(t)) * 2.5 #add some noise
```

```python
plt.plot(t, signal)
```

```python
# Perform the FFT
X = np.fft.fft(signal)

# Get the frequency bins and shift them for a centered plot
frequencies = np.fft.fftfreq(100, d=t[1]-t[0])
shifted_frequencies = np.fft.fftshift(frequencies)
shifted_X = np.fft.fftshift(X)

# Plot the magnitude spectrum
plt.figure(figsize=(8, 4))
plt.plot(shifted_frequencies, np.abs(shifted_X))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Magnitude Spectrum')
plt.xticks(np.linspace(-0.5, 0.5, 11))
plt.grid(True)
plt.show()
```
